```python
#### RG PLOTTING ####
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load data
file_path = "C:/Users/user/Desktop/ART_resistance_project/ART_resistance_project_files/new/MD_results/gyrate_files_updated/merged_gyrate_data.csv"
df = pd.read_csv(file_path)

# Set up visualization aesthetics
sns.set_style("whitegrid")
sns.set(font_scale=2)

# Define color palette for systems
palette = {
    "Wild": "black", "A675V": "purple", "C580Y": "pink", "M476I": "red", 
    "P553L": "brown", "P574L": "blue", "R561H": "green", "A557S": "orange", 
    "A578S": "cyan", "Y493H_C580Y": "yellow"
}

# Create plot
plt.figure(figsize=(10, 6))
ax = sns.violinplot(x="Systems", y="gyrate (nm)", data=df, scale="area", width=0.8, palette=palette)

# Customize axis labels and ticks
ax.set_xticklabels(ax.get_xticklabels(), size=10, weight='bold')
ax.set(yticks=[0.0, 0.4, 0.8, 1.2, 1.6], ylim=(0, 1.6))
ax.set_title("Rg of Systems")

# Optimize layout and save the figure
plt.tight_layout()
save_path = "C:/Users/user/Desktop/ART_resistance_project/ART_resistance_project_files/new/MD_results/gyrate_files_updated/rg_graph.png"
plt.savefig(save_path, format='png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
#### RMSF HEAT ####
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load data and transpose
file_path = "/home/new/Documents/Nigel/new/MD_results/rmsf_files/rmsf_line.csv"
df = pd.read_csv(file_path, sep=',').T
df.columns += 1  # Adjust column indexing to start from 1

# Set up visualization aesthetics
sns.set_style("ticks")
sns.set_style({'font.family': 'Arial'})  # Set font style

# Define color palette for y-axis labels
palette = {
    "WT": "black", "A675V": "purple", "C580Y": "pink", "M476I": "red", 
    "P553L": "brown", "P574L": "blue", "R561H": "green", "A557S": "orange", 
    "A578S": "cyan", "Y493H_C580Y": "yellow"
}

# Create plot
plt.figure(figsize=(25, 8))
heat_map = sns.heatmap(df, cmap="RdBu_r", yticklabels=1, xticklabels=4)

# Customize axis labels and ticks
heat_map.set_yticklabels(heat_map.get_yticklabels(), weight='bold', rotation=0, fontdict={"fontsize": 20})
heat_map.set_xticklabels(heat_map.get_xticklabels(), weight='bold', rotation=90, fontdict={"fontsize": 20})

# Apply color palette to y-axis tick labels
for tick_label in heat_map.get_yticklabels():
    tick_text = tick_label.get_text()
    tick_label.set_color(palette.get(tick_text, "black"))  # Use default black if key not found

# Optimize layout and save the figure
plt.tight_layout()
save_path = "/home/new/Documents/Nigel/new/MD_results/rmsf_files/rmsf_heat.png"
plt.savefig(save_path, format='png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
#### RMSF LINE ####
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load data
file_path = "/home/new/Documents/Nigel/new/MD_results/rmsf_files/rmsf_line.csv"
df = pd.read_csv(file_path, sep=',')

# Calculate mean and standard deviation
mean = df.mean().mean()
sd = pd.Series(df.values.ravel()).std()

# Set up the plot aesthetics
plt.figure(figsize=(17, 5))
sns.set_style("ticks")

# Define color palette for each system
palette = {
    "WT": "black", "A675V": "purple", "C580Y": "pink", "M476I": "red",
    "P553L": "brown", "P574L": "blue", "R561H": "green", "A557S": "orange",
    "A578S": "cyan", "Y493H_C580Y": "yellow"
}

# Plot each system's line
for system, color in palette.items():
    sns.lineplot(data=df[system], color=color, label=system)

# Add horizontal lines for mean and mean  2*SD
plt.axhline(mean, color="black", linestyle='--', label='Mean')
plt.axhline(mean + 2*sd, color="red", linestyle='--', label='Mean + 2*SD')

# Customize labels and title
plt.xlabel("Residue", fontsize=12)
plt.ylabel("Time (nm)", fontsize=12)
plt.title("RMSF", fontsize=14)

# Position the legend outside the plot
plt.legend(bbox_to_anchor=(1.01, 1), loc="upper left", borderaxespad=0)

# Optimize layout and save the figure
plt.tight_layout()
save_path = "/home/new/Documents/Nigel/new/MD_results/rmsf_files/rmsf_line2.png"
plt.savefig(save_path, format='png', dpi=600, bbox_inches='tight')

plt.show()
```
